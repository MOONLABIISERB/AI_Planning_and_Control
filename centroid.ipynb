```python
def get_centroid(x,y):
    centroid = (sum(x)/len(x),sum(y)/len(y))
    return centroid
    
def cart2pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    return(rho, phi)

def pol2cart(rho, phi):
    x = rho * np.cos(phi)
    y = rho * np.sin(phi)
    return(x, y)

def createLane(x, y,r_ = 1.01):
    cent = get_centroid(x, y)
    x_ = []
    y_ = []
    for i in range(len(x)):
        x_.append(x[i]-cent[0])
        y_.append(y[i]-cent[1])
    rad = []
    theta = []
    for i in range(len(x_)):
        r, phi = cart2pol(x_[i], y_[i])
        rad.append(r)
        theta.append(phi)
    # Scale the radius here
    rad_ = []
    for r in rad:
        rad_.append(r*r_)

    X = []
    Y = []
    for i in range(len(rad_)):
        X_, Y_ = pol2cart(rad_[i], theta[i])
        X.append(X_+cent[0])
        Y.append(Y_+cent[1])
    return np.array(X), np.array(Y)

```
